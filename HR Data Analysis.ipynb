```python
df = hr_analysis
df['Attrition'] = df['Attrition'].replace({'Yes': 1, 'No': 0})
```

Attrition Analysis

```python
# Overall Attrition Rate
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print("Attrition Rate (%):\n", attrition_rate)

# Attrition by Department
attrition_by_dept = df.groupby('Department')['Attrition'].mean() * 100
print("\nAttrition by Department (%):\n", attrition_by_dept)

# Attrition by Job Role
attrition_by_role = df.groupby('JobRole')['Attrition'].mean() * 100
print("\nAttrition by Job Role (%):\n", attrition_by_role)
```

Salary and Income Analysis

```python
# Distribution of Monthly Income
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title('Monthly Income Distribution')
plt.show()

# Income by Job Level
sns.boxplot(x='JobLevel', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Job Level')
plt.show()
```

Work-Life Balance and Satisfaction

```python
# Job Satisfaction vs Attrition
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title('Job Satisfaction and Attrition')
plt.show()

# Work-Life Balance distribution
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)
plt.title('Work-Life Balance vs Attrition')
plt.show()
```

Promotion and Career Growth Analysis

```python
# Impact of Total Working Years
sns.boxplot(x='Attrition', y='TotalWorkingYears', data=df)
plt.title('Total Working Years and Attrition')
plt.show()

# Years Since Last Promotion
sns.histplot(df['YearsSinceLastPromotion'], kde=True)
plt.title('Years Since Last Promotion and Attrition')
plt.show()
```

Travel and Distance Analysis

```python
# Business Travel and Attrition
sns.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.title('Business Travel vs Attrition')
plt.show()

# Distance from Home
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df)
plt.title('Distance from Home and Attrition')
plt.show()
```

 Overtime and Performance Impact

```python
# Overtime Analysis
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('OverTime vs Attrition')
plt.show()

# Performance Ratings
sns.countplot(x='PerformanceRating', hue='Attrition', data=df)
plt.title('Performance Rating vs Attrition')
plt.show()
```

```python
# Compute correlation only for numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(15,10))
sns.heatmap(numeric_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
